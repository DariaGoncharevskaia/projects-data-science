NaN огромное количество.
Найдем процент содержания NaN в каждом столбце относительно количества строк.

```python
def percent(df, old_threshold=None):
    nulls_percent = (df.isnull().sum()/len(df)*100).sort_values().round(1)
    threshold = nulls_percent.mean()
    a = nulls_percent.plot(kind='barh', figsize=(17, 8), title='NaN % (23699 строк)', color='#C0C0C0', legend=False, fontsize=14)
    a.set_xlabel('Count of NaN')
    dict_percent = dict(nulls_percent)
    i = 0
    for k in dict_percent:
        color = '#008000'
        if dict_percent[k] > 0:
            if dict_percent[k] > threshold:
                color = '#8B0000'
            a.text(dict_percent[k]+0.1, i + 0.09, str(dict_percent[k])+'%', color=color, 
                    fontweight='bold', fontsize=12)
            
            
        i += 0.98
    if old_threshold is not None:
        plt.axvline(x=old_threshold, linewidth=1, color='#8B0000', linestyle='--')
        a.text(old_threshold, 5, '{0:.2%}'.format(old_threshold/100), color='#8B0000', fontweight='bold', fontsize=14)
        plt.axvline(x=threshold,linewidth=1, color='#008000', linestyle='--')
        a.text(threshold, 5, '{0:.2%}'.format(threshold/100), color='#008000', fontweight='bold', fontsize=14)
    else:
        plt.axvline(x=threshold,linewidth=1, color='#8B0000', linestyle='--')
        a.text(threshold, 5, '{0:.2%}'.format(threshold/100), color='#8B0000', fontweight='bold', fontsize=14)
    return a, threshold


plot, threshold = percent(realty)
```

- 5-18 км - квартира на Парнасе или в Купчино. 
- больше 18 км - за КАДом.

В задании указано, что в датафрейм необходимо добавить некоторые новые столбцы: цена кв.м.; год, месяц, день недели публикации объявления; категории по этажам (первый, последний, другой); сотношение жилой площади и площади кухни с общей площадью.

Поэтому для начала исправим столбец first_day_exposition, после чего добавим новые столбцы, и далее изменим все типы данных. 

Это поможет нам избежать ситуации, в которой тип данных придется менять дважды: для уже существующих столбцов сейчас, и для новых чуть позже.

Меняем формат 'first_day_exposition' на datetime.

```python
realty['first_day_exposition'] = pd.to_datetime(realty['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

Добавляем новые столбцы.

```python
#посчитаем и добавим цену квадратного метра
realty['price_of_m2'] = realty['last_price']/realty['total_area']

#год, месяц, день недели публикации объявления:

realty['year'] = realty['first_day_exposition'].dt.year
realty['month'] = realty['first_day_exposition'].dt.month
realty['weekday'] = realty['first_day_exposition'].dt.weekday

#создадим словарь, для изменения чисел дней недели на текст:
dict_of_weekdays = {0: '0 - понедельник', 1: '1 - вторник', 2: '2 - среда', 3: '3 - четверг', 4: '4 - пятница', 5: '5 - суббота', 6: '6 - воскресенье'}
#меняем:
realty['weekday'] = realty['weekday'].map(dict_of_weekdays)

#категоризация по этажам (варианты — 'первый', 'последний', 'другой'):
realty['level'] = np.where(realty['floor'] == 1, 'первый', 'другой')
realty['level'] = np.where(realty['floor'] == realty['floors_total'], 'последний', realty['level'])

#столбцы с соотношением жилой и общей площади, а также с отношением площади кухни к общей:
realty['living_area_part'] = realty['living_area']/realty['total_area']
realty['kitchen_area_part'] = realty['kitchen_area']/realty['total_area']
```

Изменение типов данных.

```python
#список столбцов для приведения к типу int:
list_to_int = ['last_price', 'floors_total', 'balcony', 'airports_nearest', 'cityCenters_nearest', 'parks_around3000', 'parks_nearest', 'ponds_around3000', 'ponds_nearest', 'days_exposition', 'price_of_m2']

#изменяем на тип int 
for column in list_to_int:
    realty[column] = realty[realty[column].notnull()][column].astype('int')

#список столбцов для округления до 1 знака после запятой:
list_to_float_1 = ['total_area', 'ceiling_height', 'living_area', 'kitchen_area']

#округляем до 1 знака после запятой:
for column in list_to_float_1:
    realty[column] = realty[realty[column].notnull()][column].round(1)

#список столбцов для округления до 2 знаков после запятой:
list_to_float_1 = ['living_area_part', 'kitchen_area_part']

#округляем до 2 знаков после запятой:
for column in list_to_float_1:
    realty[column] = realty[realty[column].notnull()][column].round(2)
```

В этой части будет много графиков, оптимизируем процесс их создания.

```python
def depend(raelty, column):
    title_dict = {'rooms': 'Зависимость цены от количества комнат', 'level': 'Зависимость цены от этажа',
                  'fromcenter_category': 'Зависимость цены от удаленности от центра города',
                  'cityCenters_nearest_km': 'Зависимость цены от удаленности от центра города',
                  }
    plt.style.use('seaborn-pastel')
    if column == 'level' or column == 'fromcenter_category':
        sort_key = 'last_price'
    else:
        sort_key = column
    if column == 'fromcenter_category':
        fontsize = 15
    else:
        fontsize = 15
    filtered = raelty.groupby(column).median()['last_price'].reset_index().sort_values(by=sort_key)
    plot = filtered.plot(x=column, y="last_price", kind='bar', ec='grey', 
                            title=title_dict[column], legend=None, 
                            fontsize=fontsize, figsize=(10,5))
    ax2 = plt.gca() 
    x_axis = ax2.xaxis
    x_axis.label.set_visible(False)
    if column != 'cityCenters_nearest_kms':
        plt.xticks(rotation=0)
    return plot
```

**Зависимость от даты размещения.**

*День недели*

```python
def dmy(realty, ylims=None):
    dmy_to_plot = ['weekday', 'month', 'year']
    titles = ['Медианное значение цен по дню недели', 'Медианное значение цен по месяцу', 'Медианное значение цен по году']

    for i in range(len(dmy_to_plot)):
        realty_to_plot = realty.groupby(dmy_to_plot[i]).median()['last_price'].reset_index().sort_values(by=dmy_to_plot[i])
        if ylims is None:
            plot = realty_to_plot.plot(x=dmy_to_plot[i], y="last_price", kind='bar', figsize=(10,5), ec='black',
                                   legend=False, title=titles[i], fontsize=15)
        else:
            plot = realty_to_plot.plot(x=dmy_to_plot[i], y="last_price", kind='bar', figsize=(10,5), ec='black',
                               legend=False, title=titles[i], fontsize=15, ylim=ylims[i])
        ax2 = plt.gca()
        x_axis = ax2.xaxis
        x_axis.label.set_visible(False)
        if dmy_to_plot[i] != 'weekday':
            plt.xticks(rotation=0)
    return plot

#сразу увеличим и ограничим график, чтобы получить данные для анализа.
dmy(realty, ylims=[(4000000, 4800000), (4000000, 4850000), (4000000, 8000000)])
```

- Зависимость от дня недели: 
Почти отсутсвует. Интересно, что во вторник цена немного выше, чем в другие дни, а в выходные выкладывают более выгодные предложения. Вынесу предположение, что в рабочие дни объявления чаще публикуют  риелторы, а в выходные, например, собственники, у которых появилось на это время в свободный день. Собственникам нет необходимости выплачивать издержки на третьих лиц, поэтому их цена немного ниже.


- Зависимость от месяца:  
Объявления, опубликованные в апреле, самые "дорогие", это может быть связано с некоторой активизацией рынка. Самые выгодные предложения публикуют в июне,- летoм pынoк мoжeт «cпaть», пocкoлькy мнoгиe пoкyпaтeли нaxoдятcя в oтпycкax, поэтому продать что-то дорого довольно трудно.


- Зависимость от года:  
Такая высокая цена в 2014 году могла быть вызвана довольно скоротечным ослаблением рубля, из-за чего на рынке был ажиотаж. По этой причине, спрос мог превышать предложение, отсюда и цены. 2015 год был периодом крайне неблагоприятного экономического положения. Западные санкции, экономическое эмбарго, нестабильное геополитическое положение, падение стоимости нефти, а также множество других факторов привели страну в состояние тяжелейшей рецессии. С этим может быть связано резкое падение цен вплоть до 2017 года почти на 30%. Но не все так плохо, после 2017 года цена начала медленно расти.

Площадь, цена, число комнат, высота потолков изучены на предмет аномалий. При исследовании время продажи квартир, было отмечено влияние автоматического снятия объявлений с Яндекс Недвижимость, эти данные не учитывались. Нормальные значения продажи квартир варьируются от 1 до 600 дней, среднее время продажи 25-50, в центре квартиры продаются быстрее. На стоимость квартиры больше всего влияют: площадь, удаленость от центра, количество комнат. Незначительно на стоимость влияют этаж, наличие парков и прудов. Можно отметить, что в 2016-2017 годах стоимость квартир в среднем по рынку была ниже на 10-20 %, а в центре Санкт-Петербурга данной зависимости от года не обнаружено. На данный момент существует тенденция к удорожанию квартир в целом.
